```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.utils import shuffle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
import math

trab = pd.read_csv("C:\\Users\\danil\\Documents\\Mestrado\\Redes Neurais\\MLP Regressor\\base.csv",sep=';')

trab = shuffle(trab.head(100))

trab.describe()
```

```python

X = trab.drop(['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11','s12'],axis=1)
y = trab.iloc[:,60]

X = MinMaxScaler().fit_transform(X)

#Train data
train_X = X[:int(len(X)*0.8)]
train_y = y[:int(len(y)*0.8)]
#Test data
test_X = X[int(len(X)*0.8):]
test_y = y[int(len(y)*0.8):]


erros = []
```

```python

nn = MLPRegressor(
    hidden_layer_sizes=(10,),  activation='logistic', solver='sgd', alpha=0.001, batch_size=200,
    learning_rate='constant', learning_rate_init=0.01, power_t=0.5, max_iter=200, shuffle=True,
    random_state=9, tol=0.0001, verbose=False, warm_start=False, momentum=0.9, nesterovs_momentum=True,
    early_stopping=False, validation_fraction=0.1)

n = nn.fit(train_X, train_y)
predictions = nn.predict(test_X)

print(test_y)

for y,p in zip(test_y,predictions):
    print(y)
    print ('esperado:'+str(y)+ ' Medido: '+str(p) )
```
